```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = {
    'Number': [1, 2, 3, 4, 5, 6],
    'State': [
        'Maharashtra', 'Karnataka', 'Gujarat',
        'Delhi', 'Madhya Pradesh', 'Kerala'
    ],
    'Active_Cases': [2000, 1278, 694, 235, 548, 1836],
    'Positive': [36, 45, 100, 150, 65, 87]
}

COVID_TESTS = pd.DataFrame(data)
COVID_TESTS
```

```python
COVID_TESTS.tail(2)
```

```python
sorted_by_positive = COVID_TESTS.sort_values(by='Positive', ascending=False).reset_index(drop=True)
sorted_by_positive
```

```python
states_ge_100 = COVID_TESTS.loc[COVID_TESTS['Positive'] >= 100, 'State'].tolist()
print("States with positive cases >= 100:")
for s in states_ge_100:
    print("-", s)
```

```python
cond = (COVID_TESTS['Positive'] < 100) & (COVID_TESTS['Active_Cases'] > 1000)
COVID_TESTS.loc[cond]
```

```python
x = range(len(COVID_TESTS))
width = 0.35

plt.figure(figsize=(10,6))
plt.bar([i - width/2 for i in x], COVID_TESTS['Active_Cases'], width=width, label='Active Cases')
plt.bar([i + width/2 for i in x], COVID_TESTS['Positive'], width=width, label='Positive Cases')

plt.xticks(ticks=x, labels=COVID_TESTS['State'], rotation=45, ha='right')
plt.xlabel('State')
plt.ylabel('Number of Cases')
plt.title('Active vs Positive Cases by State')
plt.legend()
plt.tight_layout()
plt.show()
```
