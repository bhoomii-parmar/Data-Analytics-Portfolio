```python
# Perform ANOVA (F-test) between 'Toll Charges Reasonability' and 'Economic Impact'
f_stat, p_value = f_oneway(
    survey_data[survey_data["Toll Charges Reasonability"] == 1]["Economic Impact (Trade, Tourism)"],
    survey_data[survey_data["Toll Charges Reasonability"] == 2]["Economic Impact (Trade, Tourism)"],
    survey_data[survey_data["Toll Charges Reasonability"] == 3]["Economic Impact (Trade, Tourism)"],
    survey_data[survey_data["Toll Charges Reasonability"] == 4]["Economic Impact (Trade, Tourism)"]
)

print(f"ANOVA Test Result: F-statistic = {f_stat:.2f}, p-value = {p_value:.5f}")
```

```python
# Create a contingency table
contingency_table = pd.crosstab(survey_data["Toll Charges Reasonability"], survey_data["Challenges Faced"])

# Perform Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Test Result: Chi2-statistic = {chi2_stat:.2f}, p-value = {p_value:.5f}")
```

```python
X = survey_data[["Toll Charges Reasonability", "Challenges Faced", "Impact on Travel Experience", "Mode of Transport", "Frequency of Usage"]]  # Independent Variables
y = survey_data["Economic Impact (Trade, Tourism)"]  # Dependent Variable

# Add constant term to independent variables (for intercept)
X = sm.add_constant(X)

# Perform OLS Regression
model = sm.OLS(y, X).fit()

# Display the OLS summary
print(model.summary())
```
