```python
import pandas as pd

# Load the uploaded file (exact file name)
df = pd.read_csv("Mutual Fund Dataset 2025.csv")

# Preview
df.head()


# Clean and standardize all column names
df.columns = (
    df.columns
    .str.strip()                       # Remove leading/trailing spaces
    .str.replace(' ', '_')            # Replace spaces with underscores
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
    .str.replace('%', 'percent')      # Replace % with the word percent
    .str.replace('.', '', regex=False)
)

# Create a working copy
data = df.copy()

# Preview column names
print(data.columns.tolist())


import numpy as np

# Clean AUM
data['AUMin_Rs_cr'] = (
    data['AUMin_Rs_cr']
    .replace(['-', 'None'], np.nan)
    .astype(str)
    .str.replace(',', '', regex=False)
    .astype(float)
)

# Clean Expense Ratio
data['ExpenseRatio_percent'] = pd.to_numeric(data['ExpenseRatio_percent'], errors='coerce')

# Clean NAV
data['NAV'] = pd.to_numeric(data['NAV'], errors='coerce')

# Convert Inception Date to datetime
data['Inception_Date'] = pd.to_datetime(data['Inception_Date'], errors='coerce', dayfirst=True)

# Clean other performance-related numeric columns
cols_to_clean = [
    'Alpha', 'Sharpe', 'Sortino', 'Beta', 'Standard_Deviation',
    'Yield_To_Maturity_percent', 'Mod_Duration_in_yrs', 'Avg_Maturity_in_yrs'
]

for col in cols_to_clean:
    if col in data.columns:
        data[col] = pd.to_numeric(data[col].replace(['-', 'None'], np.nan), errors='coerce')

data[['Funds', 'AUMin_Rs_cr', 'ExpenseRatio_percent', 'NAV']].head()

#Stage 1 (advanced analysis)
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot style
sns.set(style="whitegrid")

# which fund categories perform best?
import numpy as np

return_cols = [
    'Return_percent1_yr', 'Return_percent3_yrs', 'Return_percent5_yrs',
    'Return_percent6_mo', 'Return_percent3_mo', 'Return_percent1_mo',
    'Return_percent2_yrs', 'Return_percent10_yrs'
]

# Convert to numeric safely
for col in return_cols:
    data[col] = pd.to_numeric(data[col].replace(['-', 'None'], np.nan), errors='coerce')


# Clean category-level returns
category_returns = data.groupby('Category')[
    ['Return_percent1_yr', 'Return_percent3_yrs', 'Return_percent5_yrs']
].mean().sort_values(by='Return_percent5_yrs', ascending=False)

# Plot
category_returns.plot(kind='bar', figsize=(14,6))
plt.title("Average Returns by Fund Category (1, 3, 5 years)")
plt.ylabel("Average Return (%)")
plt.xlabel("Fund Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Which Fund Managers Have the Best Track Records?

top_managers = data.groupby('Fund_Manager')['Return_percent3_yrs'] \
                   .mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_managers.values, y=top_managers.index, palette='rocket')
plt.title("Top 10 Fund Managers by 3-Year Avg Return")
plt.xlabel("Average 3-Year Return (%)")
plt.ylabel("Fund Manager")
plt.tight_layout()
plt.show()

# What is the relationship between Risk and Return?

plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='Beta', y='Sharpe', hue='Category', alpha=0.7)

plt.title("Risk vs. Return (Sharpe vs. Beta)")
plt.xlabel("Beta (Risk)")
plt.ylabel("Sharpe Ratio (Return per Unit Risk)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Convert risk/return ratios to numeric
data['Sharpe'] = pd.to_numeric(data['Sharpe'].replace(['-', 'None'], np.nan), errors='coerce')
data['Beta'] = pd.to_numeric(data['Beta'].replace(['-', 'None'], np.nan), errors='coerce')

# Does Higher AUM Mean Better Returns?

data['AUMin_Rs_cr'] = pd.to_numeric(data['AUMin_Rs_cr'].replace(['-', 'None'], np.nan), errors='coerce')
data['Return_percent3_yrs'] = pd.to_numeric(data['Return_percent3_yrs'], errors='coerce')

plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='AUMin_Rs_cr', y='Return_percent3_yrs', hue='Category', alpha=0.6)

plt.title("AUM vs. 3-Year Return")
plt.xlabel("Assets Under Management (₹ Cr)")
plt.ylabel("3-Year Return (%)")
plt.tight_layout()
plt.show()


#Does a Higher Expense Ratio Mean Lower Returns?
data['ExpenseRatio_percent'] = pd.to_numeric(data['ExpenseRatio_percent'].replace(['-', 'None'], np.nan), errors='coerce')
data['Return_percent5_yrs'] = pd.to_numeric(data['Return_percent5_yrs'], errors='coerce')

plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='ExpenseRatio_percent', y='Return_percent5_yrs', hue='Category', alpha=0.6)

plt.title("Expense Ratio vs. 5-Year Return")
plt.xlabel("Expense Ratio (%)")
plt.ylabel("5-Year Return (%)")
plt.tight_layout()
plt.show()

#modelling and strategy
#Prepare the Data for Clustering
cluster_data = data[['Sharpe', 'Beta', 'Standard_Deviation', 'Return_percent3_yrs']].copy()

# Convert to numeric
for col in cluster_data.columns:
    cluster_data[col] = pd.to_numeric(cluster_data[col], errors='coerce')

# Drop missing values
cluster_data.dropna(inplace=True)

# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_data)

#Run K-Means Clustering

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
k_range = range(1, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(k_range, inertia, marker='o')
plt.title("Elbow Method - Optimal k")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()

#Apply KMeans with the optimal k

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

# Add cluster labels back to original dataset
data_clustered = data.copy()
data_clustered = data_clustered.loc[cluster_data.index]
data_clustered['Cluster'] = clusters

#Visualize the Clusters

import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=cluster_data['Beta'],
    y=cluster_data['Sharpe'],
    hue=clusters,
    palette='Set1'
)
plt.title("Clusters of Mutual Funds (Risk vs Return)")
plt.xlabel("Beta (Risk)")
plt.ylabel("Sharpe Ratio (Return)")
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()

#Interpretation: What Do Your Clusters Mean?

# Check mean values of each cluster
data_clustered.groupby('Cluster')[['Sharpe', 'Beta', 'Standard_Deviation', 'Return_percent3_yrs']].mean()

cluster_labels = {
    0: 'Conservative',
    1: 'Balanced',
    2: 'Aggressive'
}
data_clustered['Profile'] = data_clustered['Cluster'].map(cluster_labels)

#Mutual Fund Recommendation System
#Define Function to Recommend Funds

def recommend_funds(profile, top_n=5):
    return data_clustered[data_clustered['Profile'] == profile][
        ['Funds', 'Category', 'Sharpe', 'Beta', 'Return_percent3_yrs']
    ].sort_values(by='Sharpe', ascending=False).head(top_n)

recommend_funds('Conservative', top_n=5)
recommend_funds('Balanced', top_n=5)
recommend_funds('Aggressive', top_n=5)

# Investor Portfolio Planning
# Define Portfolio Allocation Strategy

portfolio = {
    'Beginner': recommend_funds('Conservative', top_n=5),
    'Balanced': recommend_funds('Balanced', top_n=5),
    'Aggressive': recommend_funds('Aggressive', top_n=5)
}

for investor_type, funds in portfolio.items():
    print(f"\n📘 Portfolio for {investor_type} Investor:\n")
    display(funds)

for investor_type, funds in portfolio.items():
    print(f"\n💰 Portfolio for {investor_type} Investor (₹1,00,000):\n")
    allocation = 100000 / len(funds)
    funds['Amount Allocated'] = round(allocation, 2)
    display(funds[['Funds', 'Amount Allocated', 'Sharpe', 'Return_percent3_yrs']])

#Adding Fund Performance Tiers

def assign_tier(row):
    if row['Sharpe'] >= 1.5:
        return 'Gold'
    elif row['Sharpe'] >= 1.0:
        return 'Silver'
    else:
        return 'Bronze'

data_clustered['Tier'] = data_clustered.apply(assign_tier, axis=1)

#interactive block

# ========== User Inputs ==========
investment_goal = "Wealth Creation"   # Options: 'Tax Saving', 'Wealth Creation', 'Retirement'
risk_profile = "Balanced"             # Options: 'Conservative', 'Balanced', 'Aggressive'
investment_amount = 100000            # Any amount (₹)

# ========== Filter Funds by Risk Profile ==========
recommended = data_clustered[data_clustered['Profile'] == risk_profile].copy()

# Apply goal-specific logic (optional enhancement)
if investment_goal == "Tax Saving":
    recommended = recommended[recommended['Category'].str.contains("ELSS", case=False)]
elif investment_goal == "Retirement":
    recommended = recommended[recommended['Category'].str.contains("Hybrid|Debt", case=False)]
# Wealth Creation includes all

# Fallback: if filtered data is too small
if recommended.shape[0] < 5:
    recommended = data_clustered[data_clustered['Profile'] == risk_profile].copy()

# Sort by Sharpe and select top 5
recommended = recommended.sort_values(by='Sharpe', ascending=False).head(5)

# Allocate Investment
recommended['Allocated Amount (₹)'] = round(investment_amount / len(recommended), 2)

# Simulate 3-Year Growth
recommended['Simulated 3Y Value (₹)'] = (
    (recommended['Return_percent3_yrs'].astype(float) / 100 + 1) ** 3
    * recommended['Allocated Amount (₹)']
).round(2)

# ========== Output ==========
print(f"📊 Recommended Portfolio for a {risk_profile} Investor | Goal: {investment_goal} | ₹{investment_amount:,}")
display(recommended[['Funds', 'Category', 'Sharpe', 'Return_percent3_yrs', 'Allocated Amount (₹)', 'Simulated 3Y Value (₹)']])

#bonus
# ===== USER INPUTS =====
investment_goal = "Wealth Creation"     # Options: 'Tax Saving', 'Wealth Creation', 'Retirement'
risk_profile = "Balanced"               # Options: 'Conservative', 'Balanced', 'Aggressive'
investment_amount = 100000              # ₹50,000 or ₹1L or any
fund_tier = "Gold"                      # Options: 'Gold', 'Silver', 'Any'
num_funds = 5                           # Choose top N funds (3, 5, 7)

# ===== FILTER FUNDS =====
recommended = data_clustered[data_clustered['Profile'] == risk_profile].copy()

# Goal-based filter
if investment_goal == "Tax Saving":
    recommended = recommended[recommended['Category'].str.contains("ELSS", case=False)]
elif investment_goal == "Retirement":
    recommended = recommended[recommended['Category'].str.contains("Hybrid|Debt", case=False)]

# Tier filter
if fund_tier != "Any" and 'Tier' in recommended.columns:
    recommended = recommended[recommended['Tier'] == fund_tier]

# Fallback if too few funds
if recommended.shape[0] < num_funds:
    recommended = data_clustered[data_clustered['Profile'] == risk_profile].copy()
    if fund_tier != "Any" and 'Tier' in recommended.columns:
        recommended = recommended[recommended['Tier'] == fund_tier]

# Sort by Sharpe and pick top N
recommended = recommended.sort_values(by='Sharpe', ascending=False).head(num_funds)

# ===== CALCULATE ALLOCATIONS AND CAGR =====
recommended['Allocated Amount (₹)'] = round(investment_amount / num_funds, 2)

# 3Y Future Value Simulation
recommended['Simulated 3Y Value (₹)'] = (
    (recommended['Return_percent3_yrs'].astype(float) / 100 + 1) ** 3
    * recommended['Allocated Amount (₹)']
).round(2)

# Expected CAGR from 3Y return
recommended['Expected CAGR (%)'] = ((recommended['Return_percent3_yrs'] / 3).round(2))

# ===== OUTPUT =====
print(f"📊 Personalized Portfolio for: {risk_profile} | Goal: {investment_goal} | Tier: {fund_tier} | ₹{investment_amount:,}")
display(recommended[['Funds', 'Category', 'Sharpe', 'Return_percent3_yrs',
                     'Expected CAGR (%)', 'Allocated Amount (₹)', 'Simulated 3Y Value (₹)']])


#statstical tools
#Z-SCORE: Spot Statistically Exceptional Funds

from scipy.stats import zscore

# Compute Z-score on Sharpe ratio (NaN-safe)
data_clustered['Sharpe_Z'] = zscore(data_clustered['Sharpe'].fillna(0))

# Outliers: Sharpe ratio above Z = 2
top_performers = data_clustered[data_clustered['Sharpe_Z'] > 2]

print(f"🔍 Found {len(top_performers)} statistically high-performing funds (Z > 2)")
display(top_performers[['Funds', 'Sharpe', 'Sharpe_Z', 'Return_percent3_yrs', 'Tier', 'Profile']])


#T-TEST: Gold vs Silver Tier

from scipy.stats import ttest_ind

gold_returns = data_clustered[data_clustered['Tier'] == 'Gold']['Return_percent3_yrs'].dropna()
silver_returns = data_clustered[data_clustered['Tier'] == 'Silver']['Return_percent3_yrs'].dropna()

t_stat, p_value = ttest_ind(gold_returns, silver_returns, equal_var=False)
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("✅ Statistically significant difference: Gold-tier funds outperform Silver.")
else:
    print("❌ No statistically significant difference found between Gold and Silver.")


#CORRELATION: What Drives Performance

selected_cols = [
    'Sharpe', 'Beta', 'Standard_Deviation',
    'Return_percent3_yrs', 'AUMin_Rs_cr', 'ExpenseRatio_percent'
]

corr_data = data_clustered[selected_cols].dropna()
corr_matrix = corr_data.corr()

# Plot correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Key Mutual Fund Metrics")
plt.tight_layout()
plt.show()


#NORMAL DISTRIBUTION: What’s the Chance of Earning X%?
import scipy.stats as stats
import numpy as np

returns = data_clustered['Return_percent3_yrs'].dropna()
mean = returns.mean()
std = returns.std()

# Probability of return > 15%
prob = 1 - stats.norm.cdf(15, loc=mean, scale=std)
print(f"🎯 Probability of return > 15% over 3 years: {prob:.2%}")

# Plot normal distribution
sns.histplot(returns, kde=True, stat="density", color='skyblue')
x = np.linspace(returns.min(), returns.max(), 100)
p = stats.norm.pdf(x, mean, std)
plt.plot(x, p, 'k', linewidth=2)
plt.axvline(15, color='red', linestyle='--', label='15% Threshold')
plt.title("Normal Distribution Fit for 3-Year Returns")
plt.legend()
plt.tight_layout()
plt.show()

#Monte carlo simulation

import numpy as np
import matplotlib.pyplot as plt

# ========== CONFIGURATION ==========
investment = 100000            # Total portfolio amount
n_years = 3                    # Forecasting for 3 years
n_simulations = 1000           # Number of simulated paths
profile = 'Balanced'          # Change to 'Aggressive' or 'Conservative' as needed

# ========== SELECT TOP FUNDS ==========
selected_funds = data_clustered[data_clustered['Profile'] == profile] \
    .sort_values(by='Sharpe', ascending=False).head(5).copy()

# ========== MONTE CARLO SIMULATION ==========
results = []

for i, row in selected_funds.iterrows():
    mean_return = row['Return_percent3_yrs'] / 3   # Annualized return %
    std_dev = row['Standard_Deviation']            # Annualized risk
    alloc = investment / len(selected_funds)       # Equal allocation

    fund_simulated = []

    for _ in range(n_simulations):
        growth = alloc
        yearly_returns = np.random.normal(loc=mean_return/100, scale=std_dev/100, size=n_years)
        for r in yearly_returns:
            growth *= (1 + r)
        fund_simulated.append(growth)

    results.append(fund_simulated)

# ========== PORTFOLIO COMBINATION ==========
portfolio_results = np.sum(results, axis=0)  # Sum of 5 fund outcomes per simulation

# ========== VISUALIZATION ==========
plt.figure(figsize=(10, 6))
plt.hist(portfolio_results, bins=60, color='lightgreen', edgecolor='black')
plt.axvline(np.mean(portfolio_results), color='blue', linestyle='--', label='Mean Outcome')
plt.axvline(np.percentile(portfolio_results, 5), color='red', linestyle='--', label='5th Percentile')
plt.axvline(np.percentile(portfolio_results, 95), color='green', linestyle='--', label='95th Percentile')
plt.title(f"Monte Carlo Simulation of 3-Year ₹{investment:,} Portfolio\nProfile: {profile}")
plt.xlabel("Portfolio Value After 3 Years (₹)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ========== OUTPUT METRICS ==========
mean_final = np.mean(portfolio_results)
p5 = np.percentile(portfolio_results, 5)
p95 = np.percentile(portfolio_results, 95)
expected_cagr = ((mean_final / investment) ** (1/n_years) - 1) * 100

print(f"💼 Mean 3Y Portfolio Value: ₹{mean_final:,.2f}")
print(f"🔻 5th Percentile (Worst-Case): ₹{p5:,.2f}")
print(f"🔺 95th Percentile (Best-Case): ₹{p95:,.2f}")
print(f"📈 Expected CAGR: {expected_cagr:.2f}%")


#Markowitz Portfolio Optimization

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ===== Select Top N Funds (Example: Balanced Profile) =====
n_funds = 5
profile = 'Balanced'

selected_funds = data_clustered[data_clustered['Profile'] == profile] \
    .sort_values(by='Sharpe', ascending=False).head(n_funds).copy()

fund_names = selected_funds['Funds'].values
returns = selected_funds['Return_percent3_yrs'].values / 3 / 100  # Annualized decimal
risks = selected_funds['Standard_Deviation'].values / 100         # Annualized risk as decimal

# ===== Simulate Portfolios =====
n_portfolios = 5000
results = []

np.random.seed(42)

for _ in range(n_portfolios):
    weights = np.random.random(n_funds)
    weights /= np.sum(weights)

    port_return = np.sum(weights * returns)
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(np.cov(returns), weights)))
    sharpe_ratio = port_return / port_volatility

    results.append([port_return, port_volatility, sharpe_ratio, *weights])

columns = ['Return', 'Risk', 'Sharpe'] + list(fund_names)
df_results = pd.DataFrame(results, columns=columns)

# ===== Find Optimal Portfolio =====
optimal = df_results.loc[df_results['Sharpe'].idxmax()]

# ===== Plot Portfolios =====
plt.figure(figsize=(10, 6))
plt.scatter(df_results['Risk'], df_results['Return'], c=df_results['Sharpe'], cmap='viridis', s=10)
plt.colorbar(label='Sharpe Ratio')
plt.scatter(optimal['Risk'], optimal['Return'], c='red', s=50, label='Optimal Portfolio')
plt.xlabel('Risk (Volatility)')
plt.ylabel('Return')
plt.title('Markowitz Portfolio Optimization (Efficient Frontier)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ===== Output Optimal Allocation =====
print("🎯 Optimal Portfolio Allocation (Max Sharpe Ratio):\n")
for fund in fund_names:
    weight = optimal[fund]
    print(f"{fund[:40]:40s} → {weight*100:.2f}%")

print(f"\n📈 Expected Return: {optimal['Return']*100:.2f}%")
print(f"📉 Risk (Std Dev):  {optimal['Risk']*100:.2f}%")
print(f"⚖️ Sharpe Ratio:    {optimal['Sharpe']:.2f}")

from google.colab import files

# Save files
data_clustered.to_csv("mutual_fund_cleaned.csv", index=False)
recommended.to_csv("recommended_funds.csv", index=False)
df_results.to_csv("markowitz_results.csv", index=False)

# Download all 3
files.download("mutual_fund_cleaned.csv")
files.download("recommended_funds.csv")
files.download("markowitz_results.csv")


```
